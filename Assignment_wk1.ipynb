Load NetworkX and matplotlib

```python
import networkx as net
import matplotlib.pyplot as plt
```

I'll use the krackhardt_kite_graph function in the NetworkX library. Documentation on the function can be found from the site: https://networkx.github.io/documentation/stable/

As described in the function's comments, the following ids correspond to these names:

    Andre=1, Beverley=2, Carol=3, Diane=4,
    Ed=5, Fernando=6, Garth=7, Heather=8, Ike=9, Jane=10.
    
So we'll load the above names into a dictionary and then load those into the graph object before we draw it out.
* Note: The documentation on the function says that Jane=10, but on closer inspection of the function, it appears Jane = 0. I may be wrong, but changing Jane to 0 cleared a KeyError.

```python
labels = {}
labels[1] = 'Andre'
labels[2] = 'Beverley'
labels[3] = 'Carol'
labels[4] = 'Diane'
labels[5] = 'Ed'
labels[6] = 'Fernando'
labels[7] = 'Garth'
labels[8] = 'Heather'
labels[9] = 'Ike'
labels[0] = 'Jane'

obj = net.krackhardt_kite_graph(create_using=None)

net.draw(obj, labels=labels)
```
